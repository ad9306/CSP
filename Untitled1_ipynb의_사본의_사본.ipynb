필수 진행

1. 지역에 따라서 선호하는 게임 장르가 다를까 라는 질문에 대답을 하셔야합니다.
2. 연도별 게임의 트렌드가 있을까 라는 질문에 대답을 하셔야합니다.
3. 출고량이 높은 게임에 대한 분석 및 시각화 프로세스가 포함되어야 합니다.

```python
import pandas as pd

#Data 불러오기
data = 'https://ds-lecture-data.s3.ap-northeast-2.amazonaws.com/datasets/vgames2.csv'
df = pd.read_csv(data, thousands=',', index_col=0)

def toN(x):
  return pd.to_numeric(x, errors='coerce')

df['NA_Sales'] = toN(df['NA_Sales'])
df['EU_Sales'] = toN(df['EU_Sales'])
df['JP_Sales'] = toN(df['JP_Sales'])
df['Other_Sales'] = toN(df['Other_Sales'])
df.dtypes
```

```python
#결측치 제거
df.isnull().sum()
```

```python
df.dropna(axis=0, how='any', inplace=True)
df.isnull().sum()
```

```python
#총 판매량 추가
df['Total_Sales'] = df['NA_Sales'] + df['JP_Sales'] + df['EU_Sales'] + df['Other_Sales']
df
```

```python
#연도값을 정수형으로 변환
df = df[df.Year > 1970]
df['Year'] = df['Year'].astype(int)
df
```

# 1. Total 판매순위(Top20)

```python
import matplotlib.pyplot as plt
import numpy as np

#Total 판매순위(Top20)
df['Sales_Rank'] = df['Total_Sales'].rank(method='first', ascending=False)
df_TSRank = df.sort_values('Sales_Rank')
df_TSRank20 = df_TSRank[:20]
df_TSRankt20 = df_TSRank20.sort_values('Sales_Rank', ascending=False)

plt.figure(figsize=(10,8))
plt.barh(df_TSRankt20['Name'], df_TSRankt20['Total_Sales'])
plt.title('Total Sales Top20', fontsize = 20)
plt.xlabel('Sales Volume', fontsize = 20)
plt.ylabel('Game Name', fontsize = 20)
plt.yticks(np.arange(len(df_TSRankt20['Name'])), df_TSRankt20['Name'], fontsize=20)
plt.show()
```

# 3. 지역별 인기게임(Top20)

  지역별 인기게임입니다. 각 지역별로 어떤 게임이 인기가 있는지 확인해보도록 하죠.

```python
#NA
df_NA = df.drop(['JP_Sales', 'EU_Sales', 'Other_Sales', 'Total_Sales', 'Sales_Rank'], axis=1)
df_NA['NA_Rank'] = df_NA['NA_Sales'].rank(method='first', ascending=False)
df_NARank = df_NA.sort_values('NA_Rank')
df_NARank20 = df_NARank[:20]
df_NARank20
```

```python
#EU
df_EU = df.drop(['JP_Sales', 'NA_Sales', 'Other_Sales', 'Total_Sales', 'Sales_Rank'], axis=1)
df_EU['EU_Rank'] = df_EU['EU_Sales'].rank(method='first', ascending=False)
df_EURank = df_EU.sort_values('EU_Rank')
df_EURank20 = df_EURank[:20]
df_EURank20
```

```python
#JP
df_JP = df.drop(['EU_Sales', 'NA_Sales', 'Other_Sales', 'Total_Sales', 'Sales_Rank'], axis=1)
df_JP['JP_Rank'] = df_JP['JP_Sales'].rank(method='first', ascending=False)
df_JPRank = df_JP.sort_values('JP_Rank')
df_JPRank20 = df_JPRank[:20]
df_JPRank20
```

```python
#Other
df_Ot = df.drop(['EU_Sales', 'NA_Sales', 'JP_Sales', 'Total_Sales', 'Sales_Rank'], axis=1)
df_Ot['Ot_Rank'] = df_Ot['Other_Sales'].rank(method='first', ascending=False)
df_OtRank = df_Ot.sort_values('Ot_Rank')
df_OtRank20 = df_OtRank[:20]
df_OtRank20
```

# 4. 지역별 인기장르(필수 4번)

```python
#지역별 인기 장르 구하기
df_LOGE = df.drop(['Name', 'Platform', 'Year', 'Publisher', 'Total_Sales', 'Sales_Rank'], axis=1)
df_G_LOGE = df_LOGE.groupby('Genre').sum()
df_G_LOGE
```

```python
#지역별 인기 장르 시각화
df_G_LOGE.plot(kind='bar', rot=0, figsize=(11, 5))
plt.title('Popular genres by region', loc='center', fontsize= 20)
plt.xlabel('Genre', fontsize= 13)
plt.ylabel('Sales', fontsize= 13)
plt.show()
```

#  5. 연도별 인기장르


```python
df_YTSG = df.drop(['Name', 'Platform', 'Publisher', 'EU_Sales', 'NA_Sales', 'JP_Sales', 'Other_Sales', 'Sales_Rank'], axis=1)
df_YTSG
```

```python
from pandas import Series, DataFrame

df_GS_YTSG = df_YTSG.groupby(['Year', 'Genre'], as_index=False).sum()
df_GS_YTSG
```

```python
df_gsytsg = df_GS_YTSG.pivot_table(values= 'Total_Sales', index='Year', columns='Genre')
df_gsytsg=df_gsytsg.fillna(0)
df_gsytsg
```

```python
df_gsytsg.plot(figsize=(15,8))
plt.title('The number of sales of genres by year', fontsize=20)
plt.xlabel('Year', fontsize=15)
```

# 7. 회사별 판매량

```python
df_PbTSG = df.drop(['Name', 'Year', 'Genre', 'Platform', 'EU_Sales', 'NA_Sales', 'JP_Sales', 'Other_Sales', 'Sales_Rank'], axis=1)
df_PbTSG
```

```python
df_PbTSG2 = df_PbTSG.groupby('Publisher').sum()
df_PbTSG2['Rank'] = df_PbTSG2['Total_Sales'].rank(method='first', ascending=False)
df_PbRank = df_PbTSG2.sort_values('Rank')
df_PbRank = df_PbRank[:5]
df_PbRank
```

```python
plt.figure(figsize=(12, 5))
sns.barplot(data=df_PbRank, x=df_PbRank.index, y='Total_Sales')
plt.title('Sales volume by Publisher Top 5', fontsize=15)
plt.xlabel('Publisher', fontsize=12)
plt.ylabel('Total_Sales', fontsize=12)
```
